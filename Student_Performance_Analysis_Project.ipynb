# Questions and Visualization

# Q1. What is the distribution of GPA amongstudents?

```python
plt.figure(figsize=(10, 5))
plt.hist(data['GPA'], bins=10, color='blue', edgecolor='black', linewidth = 2)
plt.title('Distribution of GPA', fontweight='bold')
plt.xlabel('GPA', fontweight='bold')
plt.ylabel('Frequency', fontweight='bold')
plt.xticks(fontweight='bold')
plt.yticks(fontweight='bold')
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)
plt.show()
```

# Q2. What is the average GPA for eachgender?

```python
average_GPA_for_each_gender = data.groupby('Gender')['GPA'].mean()
```

```python
print(average_GPA_for_each_gender)
```

# Q3. How does parental education level affect studentGPA?

```python
labels = {
    0: 'None',
    1: 'High School',
    2: 'Some College',
    3: 'Bachelors',
    4: 'Higher'
}

data['ParentalEducation1'] = data['ParentalEducation'].replace(labels)

average_GPA_for_PEd = data.groupby('ParentalEducation1')['GPA'].mean().sort_values().reset_index()

print(average_GPA_for_PEd)
print('\n')
print('\n')
data[['ParentalEducation','GPA']].corr()
```

```python
label_GPA = []
for i in average_GPA_for_PEd['GPA']:
    label_GPA.append(i)
label_GPA = [round(gpa, 2) for gpa in label_GPA]

plt.figure(figsize=(14,8))
plot = plt.bar(average_GPA_for_PEd['ParentalEducation1'],average_GPA_for_PEd['GPA'], edgecolor='black', linewidth = 2)
plt.bar_label(plot, labels= label_GPA, label_type='edge', fontweight='bold' )
plt.title('Average GPA by Parental Education Level', fontsize = 16, fontweight = 'bold')
plt.xlabel('Education Level', fontsize = 14, fontweight = 'bold')
plt.ylabel('Average GPA', fontsize = 14, fontweight = 'bold')
plt.xticks(fontweight= 'bold')
plt.yticks(fontweight='bold')
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)
plt.tight_layout()
plt.show()
```

# Q4. Is there a correlation between study timeandGPA?

## 6. Conclusion and Insights

### Key Findings
Our analysis yielded several key insights:

1. GPA among students is right skew indicating more students at higher GPAs.
2. The average GPA for male students is 1.92, while for female students it is 1.89.
3. Students whose parents have a 'High School' education level have the highest average GPA of 1.94, followed by 'Some College' with 1.93, 'None' with 1.89, 'Higher' education with 1.82, and 'Bachelors' degree with 1.81.
4. There is a weak positive correlation (0.179) between study time and GPA, suggesting that as study time increases, GPA tends to slightly increase.
5. There is a strong negative correlation (-0.919) between the number of absences and GPA, indicating that as absences increase, GPA significantly decreases.
6. Student study time tends to cluster around 3-5 hours weekly, with a larger portion dedicating less time and a smaller portion studying for significantly longer.



- Parental education and weekly study time are strong predictors of student success.
- Absences have a significant negative impact on academic performance.

### Recommendations
Based on these findings, we recommend:
- Implementing support programs for students with low parental education levels.
- Implement strategies to improve attendance, as it has the strongest negative impact on GPA.
- Offer support programs or resources to encourage increased study time, particularly for students dedicating less than 3 hours weekly.
